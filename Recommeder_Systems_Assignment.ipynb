```python
books_ratings=pd.crosstab(index=books_1.UserID,columns=books_1['BookTitle'],values='BookRating',aggfunc='count')
```

```python
books_ratings
```

```python
books_ratings.fillna(value=0,inplace=True)
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(books_ratings.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```
